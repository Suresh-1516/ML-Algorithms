```python
import pandas as pd
```

```python
df = pd.read_csv('placement.csv')
```

```python
df
```

```python
data = df[['cgpa','iq','placement']]
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(data[['cgpa','iq']],data['placement'],test_size=0.1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
pridicted = model.predict(xtest)
```

```python
pridicted
```

```python
ytest
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,pridicted)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(xtrain, ytrain.values, clf=model, legend=2)
```
